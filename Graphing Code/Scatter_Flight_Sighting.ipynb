```python
%matplotlib inline

#Dependencies and set-up
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#File to Load
sightings_data = "Sightings_Duplicate-data.csv"
flights_data = "Airport-data_google.csv"
```

```python
#Read the data
sightings_data = pd.read_csv(sightings_data)
flights_data = pd.read_csv(flights_data)
```

```python
sightings_data = sightings_data[sightings_data.year > 2002]
flights_data = flights_data[flights_data.Year < 2017]
```

```python
yearly_sightings = sightings_data.groupby('year').size().reset_index(name='Sightings')
yearly_sightings
```

```python
yearly_flights = flights_data.groupby('Year').sum().reset_index()
yearly_flights
```

```python
yearly_flights['Flights'] = yearly_flights['Number of Departures'] + yearly_flights['Number of Connections']
yearly_flights
```

```python
Year_sightings = yearly_sightings['year']
total_sightings = yearly_sightings['Sightings']
# plt.scatter(Year, total_sightings, s=200, color='orange', edgecolors="black", alpha = 0.75, marker="X")
# plt.xlabel('Year')
# plt.ylabel('Total Sightings')
# plt.title('Sightings in Years')

# plt.savefig("Yearly_Sightings.png")

# plt.show()

Year = yearly_flights['Year']
total_flights = yearly_flights['Flights']
plt.scatter(Year, total_flights, s=total_sightings*10, color='red', edgecolors="black", alpha = 0.75, marker="o")
plt.xlabel('Year')
plt.ylabel('Total Flights')
plt.title('Flights in Years')
plt.savefig("Yearly_Flights(Sightings str).png")
plt.show()
```

```python
Year = yearly_flights['Year']
total_flights = yearly_flights['Flights']
total_sightings = yearly_sightings['Sightings']
plt.scatter(Year, total_flights, s=50, color='blue', edgecolors="black", alpha = 0.75, marker="^")
plt.scatter(Year, total_sightings, s=50, color='orange', edgecolors="black", alpha = 0.75, marker="X")
plt.xlabel('Year')
plt.ylabel('Flights in Years')
plt.title('Flights and Sightings in Years')
legend = plt.legend(fontsize = 8, title= "Counts", loc="best") 
plt.grid()

plt.savefig("Yearly_Flights&Sightings.png")

plt.show()
```
